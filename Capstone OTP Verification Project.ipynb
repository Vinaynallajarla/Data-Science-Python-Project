# Importing libraries


```python
import random
import smtplib
from email.message import EmailMessage
```

# Function to generate a 6-digit OTP

```python
def otp_generation():
  otp = ''.join([str(random.randint(0,9)) for i in range(6)])
  return otp
```

# Function to send the OTP via email and SMTP server setup



```python
def send_otp_via_email(otp, to_email):
  from_email = 'vinaylakshman30@gmail.com' # enter sender email
  email_password  = 'vfwz zhhh kmbx qdvz' # enter email's app password

  msg = EmailMessage()
  msg['Subject'] = "OTP verification"
  msg['From'] = from_email
  msg['To'] = to_email
  msg.set_content(f"Your OTP is: {otp}")

  try:
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(from_email, email_password)
    server.send_message(msg)
    server.quit()
    return "Email sent successfully."
  except Exception as e:
    return f"Failed to sent email: {e}"
```

# Function to verify the OTP with retry logic

```python
def otp_verification_system(to_email, input_otp):
  global otp, attempts
  if attempts == 0:
    return "No more attempts left. Please request a new OTP."

  if input_otp == otp:
    return "OTP_Verified"
  else:
    attempts -= 1
    return f"Invalid OTP. You have {attempts} attempts left."


def request_otp(to_email):
  global otp, attempts
  attempts = 3
  otp = otp_generation()
  return send_otp_via_email(otp, to_email)

# Global variables to store OTP and attempts
otp = ''
attempts = 0
```
